#Limpieza y manipulacion de datos

```python
# Cuantos valores faltantes hay por columna?
who.isna().sum()
```

```python
# Explorar los valores faltantes en la columna iso2
who[who.iso2.isna()]
```

```python
# Los valores NA se debe a que el iso2 de Namibia es "NA", y pandas lo tomo como un valor faltante
who.loc[who.country == "Namibia", "iso2"] = "NA"
```

```python
# Las demas columnas de casos con valores faltantes van a tener un valor default de 0
who.fillna(0, inplace=True)
```

```python
# Cuantos valores falantes hoy por columna en la tabla de pop?
pop.isna().sum()
```

```python
# Como puedo convertir columnas en filas utilizando Pandas?:: Existe la funcion melt()
# Buscar la funcion melt en la documentacion oficial
# Tengo una tabla de esta estructura:
# country|year|col1|col2
# Namibia|2000|100|200
# Quiero conseguir la siguiente estructura
# country|year|columnas|valor
# Namibia|2000|col1|100
# Namibia|2000|col2|200
who2 = who.melt(id_vars=["country", "iso2", "iso3", "year"])
who2.head(10)
```

```python
# Crear una conlumna que contenga el genero (Masculino >> Femenino)
# Crear una funcion de python que contenga las reglas
def asignar_genero(valor):
  if "m" in valor:
    return "Masculino"
  else:
    return "Femenino"
```

```python
who2["genero"] = who2["variable"].apply(asignar_genero)
who2.head(15)
```

```python
# the method of diagnosis (rel relapsi, sn = negative pulmonary smear, sp = pos)
def asignar_metodo(valor):
  if"rel" in valor:
    return "relapse"
  elif "sn" in valor:
    return "negative pulmonary smear"
  elif "sp" in valor:
    return "positive pulmonary smear"
  elif "ep" in valor:
    return "extrapilmonary"
  else:
    return "to define"
```

```python
who2["metodo"] = who2["variable"].apply(asignar_metodo)
who2.head()
```

```python
# age group (014 = 0-14 yrs of age, 1524 = 15-24, 2534 = 25-34, 3544 = 35-44 years of age, 4554 = 45-54, 5564 = 55-64, 65 = 65 years or older).
def asignar_grupoedad(valor):
  if "014" in valor:
    return "0-14"
  elif "1524" in valor:
    return "15-24"
  elif "2534" in valor:
    return "25-34"
  elif "3544" in valor:
    return "35-44"
  elif "4554" in valor:
    return "45-54"
  elif "5564" in valor:
    return "55-64"
  else:
    return "65+"
```

```python
who2["grupoedad"] = who2["variable"].apply(asignar_grupoedad)
who2.head()
```

```python
# remover la columna variable
who2.drop(columns=["variable"], inplace=True)
```

```python
# renombrar la columna value a casos
who2.rename(columns={"value":"cases"}, inplace=True)
```

```python
# Traer el dato de la poblacion total
df = pd.merge(who2, pop, on=["country", "year"], how="inner")
df.head(10)
```
